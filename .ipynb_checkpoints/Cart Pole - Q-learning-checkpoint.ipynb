# Cart Pole - Q-learning

```python
import gym
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import KBinsDiscretizer
import random
import math
from IPython.display import clear_output
```

```python
env = gym.make("CartPole-v0").env
```

```python
#Convert CartPole continous state space to the discrete one
n_bins = (4,8)
lower_bound = [ env.observation_space.low[2], -math.radians(50) ]
upper_bound = [ env.observation_space.high[2], math.radians(50) ]

def bins_discretize(_ , __ , angle, pole_velo):    
    bins = KBinsDiscretizer(n_bins = n_bins, encode='ordinal', strategy ='uniform')
    bins.fit([lower_bound,upper_bound])
    return tuple(map(int,bins.transform([[angle, pole_velo]])[0]))
```

```python
#Create Q-table
q_table = np.zeros(n_bins + (env.action_space.n,))
```

```python
def lr(e):
    return float(max(0.01, min(1, 1 - math.log10((e+1)/25))))
def epsilon(e):
    return float(max(0.1,min(1,1 - math.log10((e+1) / 25))))
```

```python
%%time
'''Training agent'''
# hyperparameters, not optimized
gamma = 0.7
lr = 0.2

total_rewards = []
for e in range(10001):
    state = bins_discretize(*env.reset())
    
    done = False
    
    total_rw = 0 
    while not done:
        if random.uniform(0, 1) < epsilon(e):
            action = env.action_space.sample() # explore
        else:
            action = np.argmax(q_table[state]) # exploit
            
        obs, rw, done, info = env.step(action) #Implement action
        
        next_state = bins_discretize(*obs) #discretize next state

        # Update q-value
        next_max = np.max(q_table[next_state])
        new_value = (1 - lr) * q_table[state][action] + lr * (rw + gamma * next_max)
        q_table[state][action] = new_value

        state = next_state #Move to next state
        
        total_rw += rw
        if total_rw==200: #Episode terminates if rewards reach 200
            break
    total_rewards.append(total_rw)
    if e % 100 == 0:
        clear_output(wait=True)
        print("Episode: {}".format(e))    

print("Training complete.")
```

```python
'''Evaluate agent performance'''
total_epochs = 0
episode = 100
#Solve 100 episode and evaluate the result
for i in range(episode):
    # Separate continuous state to discrete value
    state = bins_discretize(*env.reset())
    
    done = False
    epoch = 0
    
    while not done:    
        #Select action
        action = np.argmax(q_table[state])
        
        #Implement action
        obs, rw, done, _ = env.step(action)
        
        #Discretize next state
        next_state = bins_discretize(*obs)
        
        state = next_state #Move to next state
        
        epoch += 1
        if epoch == 200:
            break

    total_epochs+= epoch
```

```python
print("Result after 100 episodes")
print(f"Average timesteps per episodes {total_epochs/100}")
```

```python
'''Visualize rewards in training'''
plt.figure(figsize=(20,10))
plt.plot(range(len(total_rewards[:2000])),total_rewards[:2000], color = "r")
plt.title("Rewards during training with Q-learning",fontsize = 20)
plt.xlabel("Episodes",fontsize = 20)
plt.ylabel("Reward",fontsize = 20)
plt.show()
```
